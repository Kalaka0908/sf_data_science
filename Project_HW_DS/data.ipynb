```python
import numpy as np 
import pandas as pd 
```

```python
churn_data = pd.read_csv('dataHW/churn.csv')
churn_data.head()
```

```python
churn_data.info()
```

```python
import plotly.express as px


churn_data ['new'] = churn_data ['Exited'].apply(lambda x: 'Ушел' if x == 1 else 'Лояльный')

fig = px.histogram(churn_data , x='new', 
                   histnorm='percent',
                   text_auto=True)
fig.update_layout(bargap=0.4)

fig.update_layout(
    title_text='9.1 Кого больше?',
    xaxis_title_text='Клиенты',
    yaxis_title_text='Процент'
  
)   
```

На этом графике мы видим процентное соотношение лояльных и ушедших клиентов.
В данном случае лояльных клиетов больше( 79.63%)

```python

filtered_by_balance = churn_data[churn_data ['Balance']>2500]
fig = px.histogram(churn_data, x='Balance',  
                   nbins=30,
                  )
fig.update_layout(
title_text = '9.2 Клиенты со счетом > 2500)',
    xaxis_title_text='Баланс',
    yaxis_title_text='Количество'
)
fig.show()
```

Судя по графику мы видем,что основной баланс от 100000 до 125000

```python
churn_data['Ушел/остался'] = churn_data['Exited'].apply(lambda x: 'Ушел' if x == 1 else "Остался")

fig = px.histogram(churn_data, x='Balance',  
                   nbins=30,
                  marginal="box",
                  color = 'Ушел/остался')
                
fig.update_layout(
title_text = '9.3 Распределение баланса',
    xaxis_title_text='Баланс',
    yaxis_title_text='Количество'
)
fig.show()
```

те кто ушел баланс был небольшой

```python
fig1 = px.histogram(
    data_frame=churn_data,
    x='Age', #ось абсцисс
    color='Exited',
    height=500, #высота
    width=700, #ширина
    histfunc='count',
    title='Распределение возраста клиентов: 1 - ушедшие, 0 - лояльные', 
    nbins = 25
)

fig2 = px.box(
    data_frame=churn_data,
    x='Age', #ось абсцисс
    color='Exited',
    height=500, #высота
    width=700, #ширина
    title='Распределение возраста клиентов: 1 - ушедшие, 0 - лояльные', 
)
fig1.show()
fig2.show()
```

На этом графике мы видим что средний возраст ушедших гораздо выше чем оставшихся

```python
fig = px.histogram(churn_data, x='Gender', 
                   histnorm='percent',
                   text_auto=True,
                  color ='new',
                  barmode='group',
             histfunc='avg')
fig.update_layout(bargap=0.2)

fig.update_layout(
    title_text='9.6 Связь с возрастом',
    yaxis_title_text='Доля'
)
```

лояльных мужчин гораздо больше чем женщин! женщины предпочитают уйти

```python
fig = px.histogram(churn_data, x='NumOfProducts', 
                   histnorm='percent',
                   text_auto=True,
                   color ='new',
                   )
fig.update_layout(bargap=0.2)

fig.update_layout(
    title_text='9.7 Отток клиентов от приобретенных услуг',
    yaxis_title_text='Доля'
 )
```

кто берет больше продуктов остается не довольный

```python
churn_data['Активный'] = churn_data['IsActiveMember'].apply(lambda x: 'Да' if x == 1 else "Нет")


fig = px.histogram(churn_data, x='Активный', 
                   histnorm='percent',
                   text_auto=True,
                  color ='new',
                  barmode='group',
             histfunc='avg')
fig.update_layout(bargap=0.4)

fig.update_layout(
    title_text='9.8 Связь с активностью клиента',
    yaxis_title_text='Доля',
    xaxis_title_text='Активный клиент?'
 )
```

активные клиенты не уходят! нужно больше стимула остаться 

```python
strana = churn_data.groupby(['Geography'], as_index=False).agg('mean')

fig = px.choropleth(
    data_frame=strana, 
    locations='Geography', 
    locationmode = "country names", 
    color="Exited", 
    hover_name='Geography',
    range_color=[0, 0.4], 
    title='9.9 Доля ушедших клиентов по странам', 
    width=800, 
    height=500, 
    color_continuous_scale=px.colors.sequential.Plasma 
)

#отображаем график
fig.show()
```

больше всего ушедших в Германии

```python
salary_score_data = churn_data.groupby(['CreditScore', 'Exited'], as_index=False)['EstimatedSalary'].median()
fig = px.scatter(
    data_frame = salary_score_data,
    x = 'EstimatedSalary',
    y='CreditScore',
    color='Exited',
    height=300,
    width=700,
    title='9.5 Зависимость крединого рейтинга от предполагаемой заработной платы'
)
fig.show()
```
